```python
# Import Seaborn
import seaborn as sns

# Import dependencies
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Add a dark grid background to all plots
sns.set_theme(style="darkgrid")
```

<h1>2 BEZIEHUNGEN VISUALISIEREN</h1>
<li>Hauptfunktion: sns.relplot</li>

<h2>2.1 SCATTERPLOTS</h2>

<li>Minimum: x & y definieren, dataset definieren </li>
<li>hue -> Variabel mit Farbe aufteilen -> kann mit "palette='...'" angepasst werden</li>
<li>style -> Variabel mit Form aufteilen</li>
<li>size -> Variabel mit GrÃ¶sse aufteilen -> kann mit "sizes='...'" angepasst werden</li>


```python
# Load dataset
tips = sns.load_dataset("tips")

# Scatterplot
sns.relplot(
    x = "total_bill", 
    y = "tip", 
    hue = "smoker", 
    style = "smoker", 
    size = "size", sizes = (15,200), 
    data = tips
);
```

<h2>2.2 LINE PLOTS</h2>

<li>Default: mean x / 95 % CI</li>
<li>CI Ã¤ndern (z.B. an SD): "ci = sd"</li>
<li>Aggregation off: "estimator = None"</li>

```python
fmri = sns.load_dataset("fmri")
sns.relplot(
    x = "timepoint", 
    y = "signal", 
    hue = "region",
    style = "event",
    kind = "line", 
    data = fmri
)
```

<h1>3 DISTRIBUTIONEN VISUALISIEREN</h1>
<li>Hauptfunktion: sns.displot</li>

<h2>3.1 HISTOGRAMME</h2>

<li>dataset und x definieren</li>
<li>Breite (binwidth = #) oder Anzahl (bins = #) der Balken entscheiden </li>
<li>schÃ¶ner: spezifisches Anzahl definieren oder "discrete = True"</li>

```python
#Choose a color palette (examples at the bottom)
sns.set_palette("Paired")

#Load dataset
penguins = sns.load_dataset("penguins")

#Plot
sns.displot(
    penguins, 
    x = "flipper_length_mm",
    binwidth = 3,
)
```

<h3>3.1.1 MULTIPLE VARIABLEN</h3>

```python
sns.displot(penguins, x = "flipper_length_mm", hue = "species")
```

```python
sns.displot(penguins, x = "flipper_length_mm", hue = "species", element = "step")
```

```python
sns.displot(penguins, x = "flipper_length_mm", hue = "species", multiple = "stack")
```

```python
sns.kdeplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
```

3.2 BIVARIATE DISTRIBUTION

```python
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", hue="species")
```

```python
sns.jointplot(
    data=penguins,
    x="bill_length_mm", y="bill_depth_mm", hue="species",
    kind="kde"
)
```

<h2>3.2 KATEGORISCHE DISTRIBUTION</h2>

```python
sns.catplot(x="total_bill", y="day", hue="sex",
            kind="violin", bw=.15, cut=0,
            data=tips)
```

```python
sns.catplot(x="day", y="total_bill", data=tips)
```
